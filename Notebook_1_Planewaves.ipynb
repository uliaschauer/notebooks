```python
from pylab import *
from ipywidgets import *
%matplotlib notebook

#initial values
a0 = 1
b0 = 0
k0 = 2

#create plot and data
x = arange(0.0, 6.2831853072, 0.001)
y_real = a0*cos(k0*x)+b0*cos(-k0*x)
y_imag = a0*sin(k0*x)+b0*sin(-k0*x)
p_real, = plot(x, y_real, lw=2, color='red', label='Real')
p_imag, = plot(x, y_imag, lw=2, color='blue', label='Imaginary')
axis([0, 6.2831853072, -2, 2])
xlabel('x')
ylabel('Wellenfunktion')
l=legend(loc='upper right')
```

```python
@interact(a=(0.0, 1.0, 0.01), b=(0.0, 1.0, 0.01), k=(0.0, 10.0))
def update(a=a0, b=b0, k=k0):
    p_real.set_ydata(a*cos(k*x)+b*cos(-k*x))
    p_imag.set_ydata(a*sin(k*x)+b*sin(-k*x))
```
